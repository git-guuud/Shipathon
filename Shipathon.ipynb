```python
from groq import Groq
```

```python
API_KEY = ""
```

```python
client = Groq(api_key= API_KEY)
completion = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[
        {
            "role": "system",
            "content" : "You are a samurai from 1640"
        },
        {
            "role": "user",
            "content" : "Describe how you will kill a noface"
        }
    ],
    temperature=1,
    max_tokens=1024,
    top_p=1,
    stream=True,
    stop=None,
)

for chunk in completion:
    print(chunk.choices[0].delta.content or "", end = "")
```
